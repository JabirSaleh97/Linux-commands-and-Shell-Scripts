```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt

def extract_lines_and_colorize(img_path, color):
    # Load and convert to grayscale
    img = Image.open(img_path).convert("L")

    # Invert: lines become white, background becomes black
    img = ImageOps.invert(img)

    # Convert to numpy array
    data = np.array(img)

    # Normalize and threshold to extract lines
    mask = data > 30  # Adjust threshold if needed

    # Create transparent image
    rgba = np.zeros((data.shape[0], data.shape[1], 4), dtype=np.uint8)
    rgba[mask] = [*color, 255]  # Apply RGB color + full opacity to lines

    return Image.fromarray(rgba, 'RGBA')

# Color-coded overlays: red, green, blue
img1_colored = extract_lines_and_colorize("CO2 TD-DFT.png", (255, 0, 0))     # Red
img2_colored = extract_lines_and_colorize("NCN SINGLET 2-OH.png", (0, 255, 0)) # Green
img3_colored = extract_lines_and_colorize("NCN TDDFT.png", (0, 0, 255))      # Blue

# Composite the overlays
base = Image.new("RGBA", img1_colored.size, (255, 255, 255, 255))
combined = Image.alpha_composite(base, img1_colored)
combined = Image.alpha_composite(combined, img2_colored)
combined = Image.alpha_composite(combined, img3_colored)

# Show result
plt.figure(figsize=(10, 8))
plt.imshow(combined)
plt.axis('off')
plt.title("Overlayed UV Spectra with Color Tints")
plt.show()
```

```
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def extract_lines_and_colorize(img_path, color):
    img = Image.open(img_path).convert("L")
    img = ImageOps.invert(img)
    data = np.array(img)
    mask = data > 30  # Threshold
    rgba = np.zeros((data.shape[0], data.shape[1], 4), dtype=np.uint8)
    rgba[mask] = [*color, 255]
    return Image.fromarray(rgba, 'RGBA')

# Apply different color codes to each image
img1 = extract_lines_and_colorize("CO2 TD-DFT (1).png", (255, 0, 0))        # Red
img2 = extract_lines_and_colorize("NCN SINGLET 2-OH (1).png", (0, 255, 0))  # Green
img3 = extract_lines_and_colorize("NCN TDDFT (1).png", (0, 0, 255))         # Blue
img4 = extract_lines_and_colorize("CO2 SINGLET.png", (255, 165, 0))         # Orange

# Combine all images
base = Image.new("RGBA", img1.size, (255, 255, 255, 255))
combined = Image.alpha_composite(base, img1)
combined = Image.alpha_composite(combined, img2)
combined = Image.alpha_composite(combined, img3)
combined = Image.alpha_composite(combined, img4)

# Compact display size
fig, ax = plt.subplots(figsize=(6, 5))  # Smaller figure
ax.imshow(combined)
ax.axis('off')
ax.set_title("Overlayed UV Spectra", fontsize=10)

# Legend
legend_patches = [
    mpatches.Patch(color='red', label='CO TD-DFT'),
    mpatches.Patch(color='green', label='NCN Singlet 2-OH'),
    mpatches.Patch(color='blue', label='NCN TDDFT'),
    mpatches.Patch(color='orange', label='CO Singlet'),
]

# Legend outside to the right, small font
fig.legend(handles=legend_patches, loc='center left', bbox_to_anchor=(1.01, 0.5), borderaxespad=0., fontsize=8)

plt.tight_layout(pad=2)
plt.show()
```

```
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def extract_lines_and_colorize(img_path, color):
    img = Image.open(img_path).convert("L")
    img = ImageOps.invert(img)
    data = np.array(img)
    mask = data > 30  # Threshold
    rgba = np.zeros((data.shape[0], data.shape[1], 4), dtype=np.uint8)
    rgba[mask] = [*color, 255]
    return Image.fromarray(rgba, 'RGBA')

# Apply color codes to each spectrum image
img1 = extract_lines_and_colorize("CO2 TD-DFT (1).png", (255, 0, 0))        # Red
img2 = extract_lines_and_colorize("NCN SINGLET 2-OH (1).png", (0, 255, 0))  # Green
img3 = extract_lines_and_colorize("NCN TDDFT (1).png", (0, 0, 255))         # Blue
img4 = extract_lines_and_colorize("CO2 SINGLET.png", (255, 165, 0))         # Orange

# Combine them
base = Image.new("RGBA", img1.size, (255, 255, 255, 255))
combined = Image.alpha_composite(base, img1)
combined = Image.alpha_composite(combined, img2)
combined = Image.alpha_composite(combined, img3)
combined = Image.alpha_composite(combined, img4)

# Plot with smaller figure and fonts
fig, ax = plt.subplots(figsize=(6, 5))
ax.imshow(combined)
ax.axis('on')  # Turn on axes to show labels

# Set axis labels with smaller font
ax.set_xlabel("Wavelength (nm)", fontsize=8)
ax.set_ylabel("Arb. Units", fontsize=8)
ax.set_title("Overlayed UV Spectra", fontsize=10)

# Remove ticks but keep small axis labels
ax.tick_params(axis='both', which='both', length=0, labelsize=6)

# Legend outside
legend_patches = [
    mpatches.Patch(color='red', label='CO TD-DFT'),
    mpatches.Patch(color='green', label='NCN Singlet 2-OH'),
    mpatches.Patch(color='blue', label='NCN TDDFT'),
    mpatches.Patch(color='orange', label='CO Singlet'),
]
fig.legend(handles=legend_patches, loc='center left', bbox_to_anchor=(1.01, 0.5), borderaxespad=0., fontsize=8)

plt.tight_layout(pad=2)
plt.show()
```
